## Investigation Overview

In this investigation, I wanted to look at video games genres and see if there were certain genres that generally had hire critic and user scores. I also wanted determine if critic scores had any correlation to user scores for these games. 

## Dataset Overview and Executive Summary

The data consisted of around 16,000 video games and attributes of genres, user scores, and critic scores. The genres consisted of 'Sports', 'Platform', 'Racing', 'Role-Playing', 'Puzzle', 'Misc', 'Shooter', 'Simulation', 'Action', 'Fighting', 'Adventure', and 'Strategy' as subattributes. There was one subattribute that was removed from the analysis due to missing information. 

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
games = pd.read_csv('video_games_clean.csv')
````

## (Average Critic Scores by Video Game Genre)
>Critic Scores in this data set have a wide range of scores, they are measured on a scale of 0 to 100. To answer the question of which genre had the best average scores, I chose a bar plot to convey the average scores in descending order from highest average rating to lowest. You can see from the graph below that role-playing games had the highest average of all the genres among critics. 


````python
# Calculate the average critic score for each genre
average_scores = games.groupby('Genre')['Critic_Score'].mean().sort_values(ascending=False).reset_index()

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(average_scores['Genre'], average_scores['Critic_Score'], color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Average Critic Score')
plt.title('Average Critic Scores by Video Game Genre')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
````

## (Average User Scores by Video Game Genre)

>User Scores in this data set also have a wide range of scores, they are measured on a scale of 0 to 10. To answer the question of which genre had the best average scores, I chose a bar plot to convey the average scores in descending order from highest average rating to lowest. You can see from the graph below that role-playing games had the highest average of all the genres among users. 

````python
# Calculate the average critic score for each genre
average_scores = games.groupby('Genre')['User_Score'].mean().sort_values(ascending=False).reset_index()

# Plotting the data
plt.figure(figsize=(10, 6))
plt.bar(average_scores['Genre'], average_scores['User_Score'], color='skyblue')
plt.xlabel('Genre')
plt.ylabel('Average User Score')
plt.title('Average User Scores by Video Game Genre')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Show the plot
plt.show()
````

## (User Scores vs Critic Scores)

> The scatter plot with a linear regression line makes it easy to compare the user scores with the critic scores. Faceting it out by genre also makes this much smoother. 

>You can see, that more often than not, user and critic scores align, but there are some outliers especially within 'Shooter', 'Action', and 'Strategy' that show the critics aren't always correct with how well a game will be received by users. 

````python
# Set the style
sns.set(style="whitegrid")

# Create the lmplot
g = sns.lmplot(data=games, x='Critic_Score', y='User_Score', hue='Genre', col='Genre', col_wrap=4, markers='.', height=4, scatter_kws={'alpha':0.5})

# Add an overall title
plt.suptitle("User Scores vs. Critic Scores", y=1.02)

# Show the plot
plt.show()
````
